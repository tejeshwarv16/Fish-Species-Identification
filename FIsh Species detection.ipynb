```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import seaborn as sns
from keras.layers import Dense, Flatten
from keras.models import Model
from keras.applications.inception_v3 import InceptionV3, preprocess_input
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import keras
```

```python
img=image.imread("file location")
plt.imshow(img)
```

```python
train_gen = ImageDataGenerator(featurewise_center= True,
                               width_shift_range=0.1,
                               height_shift_range=0.1,
                               shear_range=0.1,
                              zoom_range=0.1,
                               horizontal_flip=True,
                               vertical_flip=True,
                               preprocessing_function=preprocess_input)
test_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
train_path="Training_Set location"
test_path="Test_Set location"

train_data=train_gen.flow_from_directory(train_path,
                                        target_size=(200,200),
                                        batch_size=150,
                                        class_mode="categorical"
                                        )                                        

val_data=test_gen.flow_from_directory(test_path,
                                        target_size=(200,200),
                                        batch_size=150,
                                        class_mode="categorical"
                                        )
```

```python
base_model = InceptionV3(input_shape=(200,200,3), include_top= False)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
X = Flatten()(base_model.output)
X = Dense(units=20, activation='sigmoid')(X)
model = Model(base_model.input, X)
model.compile(optimizer='adam', loss = keras.losses.binary_crossentropy, metrics=['accuracy'])
model.summary
```

```python
train_data.class_indices
```

```python
t_img , label = train_data.next()
```

```python
def plotImages(img_arr, label):
    for idx ,img in enumerate(img_arr):
        if idx <= 10:
            plt.figure(figsize=(5,5))
            plt.imshow(img)
            plt.title(img.shape)
            plt.axis=False
            plt.show()
```

```python
plotImages(t_img, label)
```

```python
from keras.callbacks import ModelCheckpoint , EarlyStopping
mc = ModelCheckpoint(filepath = "./best_model.h5",
                    mointor='accuracy',
                    verbose=1,
                    save_best_only=True)
es = EarlyStopping(monitor= "accuracy", min_delta=0.01, patience=5,
                  verbose=1)
cb = [mc,es]
```

```python
history = model.fit(
    train_data,
    batch_size=150,
    epochs=10,
    validation_data=val_data,
)
```
